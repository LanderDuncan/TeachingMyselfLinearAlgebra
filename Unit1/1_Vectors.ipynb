### Adding Vectors
#### Tip to tail method
In the below example the two vectors (red and blue) are added together by adding the x and y componenent of each vector. The graphical representation below shows that we essentially travel from the origin to the tip of one vector and then do the same for the next vector to get the resultant vector.<br> **How much did you shift in total relative to the origin.** <br>
The trasative property still applies: $$ \vec a+ \vec b= \vec b + \vec a$$
<br>
(Note: the graphical library does not show the red vector as touching the other two vectors, but it should be).
#### Paralelagram method
When both two vectors are drawn from the same point you can form a parallelagram by translating the vector onto the tip of another vector as shown by the dotted lines. From here the resultant vector is represented by the diagnal line.
![Sum.jpg](../Sum.jpg)

```python
vectorA = np.array([6,-2])
vectorB = np.array([-4,4])
print(np.add(vectorA,vectorB))

plt.quiver([0], [0], vectorA[0], vectorA[1], color='b', units='xy', scale=1)
plt.quiver([6], [-2], vectorB[0], vectorB[1], color='r', units='xy', scale=1)
plt.quiver([0], [0], [2], [2], color='purple', units='xy', scale=1)
  

plt.xlim(-10, 10)
plt.ylim(-10, 10)
  

plt.grid()
plt.show()
```

### Multiplying vectors by a scalar
To multiply a vector by a scalar you mutliply each quantity in the vector by that amount.

$$\begin{bmatrix}
2 * multiple\\
1 * multiple
\end{bmatrix}$$

If you want to multiply the vector by three you would get the following vector:
$$\begin{bmatrix}
6\\
3
\end{bmatrix}$$

Mutliplying by a scalar will not change the direction (unless you flip the sign), but the magnitude will increase by the multiple.

```python
vectorC = np.array([2,1])
print(np.linalg.norm(vectorC))
# This vector is multiplied by 3 and has a 3x greater magnitude.
vectorD = vectorC * 3
print(np.linalg.norm(vectorD))

# The direction of the vector has not changed, since the sign has not changed.
plt.quiver([0], [0], vectorD[0], vectorD[1], color='r', units='xy', scale=1)
plt.quiver([0], [0], vectorC[0], vectorC[1], color='b', units='xy', scale=1)


plt.xlim(-10, 10)
plt.ylim(-10, 10)
  

plt.grid()
plt.show()
```

### Subtracting vectors

Subtracting vectors is the same as adding the negative version of the second vector. 
$$ \vec a- \vec b = \vec a + -1\vec b$$

```python
vectorA = np.array([2,3])
vectorB = np.array([-4,-2])
print(np.subtract(vectorA,vectorB))

plt.quiver([0], [0], vectorA[0], vectorA[1], color='b', units='xy', scale=1)
plt.quiver([0], [0], vectorB[0], vectorB[1], color='r', units='xy', scale=1)
plt.quiver(vectorB[0], vectorB[1], [6], [5], color='purple', units='xy', scale=1)
  

plt.xlim(-10, 10)
plt.ylim(-10, 10)
  

plt.grid()
plt.show()
```

### Unit vectors
This is a way to represent vectors as two single unit vectors that are multiplied by a scalar. The two vectors are $$\hat i$$ which represents one horizontal unit and $$\hat j$$ which represents one vertical unit.<br> An example uses the following vector:
$$\begin{bmatrix}
2\\
3
\end{bmatrix}$$
This vector would simplify to $$ 2\hat i + 3\hat j $$
Addition in this notation works the same way, but instead of adding numbers in a collumn you add the coeficients in front of the unit vectors for each unit vector.

### Parametic representations of lines
This is a more general way to define a line that will be useful later since y = mx + b is R<sup>2</sup> specific.
Esentially if you take one vector and multiply it by some contant c you can move the magnitude infinetly creating a line with the same slope as the inital vector. In order to change the y intercept you add another vector to it that shifts the equastion.

$$ L = \{\vec x + c\vec v | c \in \mathbb{R}\} $$

#### How to find the equastion of a line that passes through two points

$$L = \{\vec a + t(\vec a - \vec b)\} $$
This equastion will give you a line that passes through two points represented by vectors a and b. The first vector can be either a or b and the vecotrs in the parenthesis can be switched to either order.
<br>
**Example:**
$$

7
\end{bmatrix}
$$
$$

3 \cr
4
\end{bmatrix}
$$
The paramentric equastion for the line will be:<br>
x = -1 + -1t <br>
y = 2 + -1t <br>
